```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame

# File to Load
file_to_load = "./Resources/MoviesOnStreamingPlatforms_updated.csv"

# Read Purchasing File and store into Pandas data frame
csv = pd.read_csv(file_to_load, engine='python')
csv.head()
```

```python
moviecount = (len(csv))
moviecount
```

```python
newdata = csv[(csv['Year'] > 2009) & (csv['Year'] < 2019)]
newdata
```

```python
total_movies=newdata["Title"].count()
total_movies
```

```python
netflix=newdata["Netflix"].value_counts()
netflix[1]
```

```python
hulu=newdata["Hulu"].value_counts()
hulu[1]
```

```python
prime_video=newdata["Prime Video"].value_counts()
prime_video[1]
```

```python
disney=newdata["Disney+"].value_counts()
disney[1]
```

```python
#create data frame
number_movies= pd.DataFrame({"Netflix": [netflix[1]], "Hulu": [hulu[1]], "Prime Video":[prime_video[1]], "Disney+":[disney[1]], "Total Movies": total_movies})
number_movies
```

```python
number_movies.plot(kind="bar", figsize=(8,5))
plt.title("Number of Movies for Each Streaming Platform")
plt.xlabel("Streaming Platforms")
plt.ylabel("Number of movies")
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Netflix', 'Hulu', 'Prime Video', 'Disney+'
sizes = [2440, 590, 6156, 135]
explode = (0, 0, 0, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 

plt.show()
```
